```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
from tensorflow.keras import layers
from keras.utils import to_categorical

img = mpimg.imread('pixelart.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
import numpy as np
import os

rootdir = 'C:\Rishi\Python\\ai\CFD Version 3.0\Images\CFD_supersmall'
x_train_original = []
x_test_original = []
y_test_original = []
y_train_original = []

AF, AM, BF, BM, LF, LM, WF, WM = (0,0,0,0,0,0,0,0)
traincount = -1
testcount = -1

testopen = open("C:\Rishi\Python\\ai\y_train.txt", "r")
testread = testopen.read()
testlines = testread.splitlines()

trainopen = open("C:\Rishi\Python\\ai\y_test.txt", "r")
trainread = trainopen.read()
trainlines = trainread.splitlines()

## NEED TO REVISIT DATA. TRAIN AND TEST SETS SHOULD HAVE EQUAL NUMBER OF EACH RACE

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        if 'N' in os.path.join(subdir, file):
            if 'AF' in os.path.join(subdir, file):
                AF = AF + 1
                if AF < 41:
                    x_train_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("AF Train Done:" + str(AF))
                    traincount = traincount + 1
                    y = round(float(trainlines[traincount]))
                    y_train_original.append(y)
                else:
                    x_test_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("AF Test Done:" + str(AF))
                    testcount = testcount + 1
                    print(testcount)
                    y = round(float(testlines[testcount]))
                    y_test_original.append(y)
            elif 'AM' in os.path.join(subdir, file):
                AM = AM + 1
                if AM < 41:
                    x_train_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("AM Train Done:" + str(AM))
                    traincount = traincount + 1
                    y = round(float(trainlines[traincount]))
                    y_train_original.append(y)
                else:
                    x_test_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("AM Test Done:" + str(AM))
                    testcount = testcount + 1
                    print(testcount)
                    y = round(float(testlines[testcount]))
                    y_test_original.append(y)
            elif 'BF' in os.path.join(subdir, file):
                BF = BF + 1
                if BF < 71:
                    x_train_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("BF Train Done:" + str(BF))
                    traincount = traincount + 1
                    y = round(float(trainlines[traincount]))
                    y_train_original.append(y)
                else:
                    x_test_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("BF Test Done:" + str(BF))
                    testcount = testcount + 1
                    print(testcount)
                    y = round(float(testlines[testcount]))
                    y_test_original.append(y)
            elif 'BM' in os.path.join(subdir, file):
                BM = BM + 1
                if BM < 71:
                    x_train_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("BM Train Done:" + str(BM))
                    traincount = traincount + 1
                    y = round(float(trainlines[traincount]))
                    y_train_original.append(y)
                else:
                    x_test_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("BM Test Done:" + str(BM))
                    testcount = testcount + 1
                    print(testcount)
                    y = round(float(testlines[testcount]))
                    y_test_original.append(y)
            elif 'LF' in os.path.join(subdir, file):
                LF = LF + 1
                if LF < 41:
                    x_train_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("LF Train Done:" + str(LF))
                    traincount = traincount + 1
                    y = round(float(trainlines[traincount]))
                    y_train_original.append(y)
                else:
                    x_test_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("LF Test Done:" + str(LF))
                    testcount = testcount + 1
                    print(testcount)
                    y = round(float(testlines[testcount]))
                    y_test_original.append(y)
            elif 'LM' in os.path.join(subdir, file):
                LM = LM + 1
                if LM < 41:
                    x_train_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("LM Train Done:" + str(LM))
                    traincount = traincount + 1
                    y = round(float(trainlines[traincount]))
                    y_train_original.append(y)
                else:
                    x_test_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("LM Test Done:" + str(LM))
                    testcount = testcount + 1
                    print(testcount)
                    y = round(float(testlines[testcount]))
                    y_test_original.append(y)
            elif 'WF' in os.path.join(subdir, file):
                WF = WF + 1
                if WF < 71:
                    x_train_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("WF Train Done:" + str(WF))
                    traincount = traincount + 1
                    y = round(float(trainlines[traincount]))
                    y_train_original.append(y)
                else:
                    x_test_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("WF Test Done:" + str(WF))
                    testcount = testcount + 1
                    print(testcount)
                    y = round(float(testlines[testcount]))
                    y_test_original.append(y)
            elif 'WM' in os.path.join(subdir, file):
                WM = WM + 1
                if WM < 71:
                    x_train_original.append(mpimg.imread(os.path.join(subdir, file)))
                    print("WM Train Done:" + str(WM))
                    traincount = traincount + 1
                    y = round(float(trainlines[traincount]))
                    y_train_original.append(y)
                else:
                    x_test_original.append(mpimg.imread(os.path.join(subdir, file)))            
                    print("WM Test Done:" + str(WM))
                    testcount = testcount + 1
                    print(testcount)
                    y = round(float(testlines[testcount]))
                    y_test_original.append(y)
                
x_train = np.array(x_train_original)
x_test = np.array(x_test_original)

y_test = np.array([y_test_original]).T
y_train = np.array([y_train_original]).T
print(str(y_train.shape) + " " + str(y_test.shape))
```

```python
model = Sequential()

model.add(Conv2D(64, (5, 5), activation='relu', input_shape=(64, 64, 3)))

model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (5, 5), activation='relu'))

model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(1000, activation='relu'))

model.add(Dropout(0.5))

model.add(Dense(500, activation='relu'))

model.add(Dropout(0.5))

model.add(Dense(250, activation='relu'))

model.add(Dense(7, activation='softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy',
              optimizer = 'adam',
              metrics = ['accuracy'])
y_train_onehot = to_categorical(y_train, 7)
y_test_onehot = to_categorical(y_test, 7)
tf.config.run_functions_eagerly(True)

print(y_train_onehot[30])
print(y_train[30])
```

```python
hist = model.fit(x_train, y_train_onehot,
                batch_size = 440,
                epochs = 5,
                validation_split = 0.1)
```

```python
model.evaluate(x_test, y_test_onehot)[1]
```
